### 1.3 Exceptions

Write `safe_divide(a, b)` that:
- Returns `a / b` normally
- Returns `None` if `b == 0` (catch `ZeroDivisionError`)
- Raises `TypeError` with message `"Inputs must be numeric"` if either input is not a number

````python
def safe_divide(a, b):
    # Safely divide two numbers, handling division by zero and type errors
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)): 
      raise TypeError("Inputs must be numeric")
    try:
      return a / b
    except ZeroDivisionError:
      return None
    pass

# Tests
assert safe_divide(10, 2) == 5.0
assert safe_divide(5, 0) is None
try:
    safe_divide('x', 2)
    assert False, "Should have raised TypeError"
except TypeError as e:
    assert str(e) == "Inputs must be numeric"
print("1.3 passed ✓")
````

### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```
A = [[4, 1],
     [2, 3]]
```

1. Compute eigenvalues and eigenvectors using `np.linalg.eig`.
2. **Verify** the eigenvector equation: `A @ v = λ * v` for each eigenpair.
3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````python
A = np.array([[4, 1], [2, 3]], dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify Av = λv for each eigenpair
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    vec = eigenvectors[:, i]
    lhs = A @ vec
    rhs = lam * vec
    assert np.allclose(lhs, rhs), f"Eigen equation failed for pair {i}"
    print(f"Verified: A @ v{i} = λ{i} * v{i}")

# Plot
fig, ax = plt.subplots()
colors = ['crimson', 'steelblue']
for i in range(len(eigenvalues)):
    lam = eigenvalues[i].real
    vec = eigenvectors[:, i].real
    scaled = vec * lam
    ax.quiver(0, 0, scaled[0], scaled[1], angles='xy', scale_units='xy', scale=1,
              color=colors[i], width=0.01, label=f'λ{i}={lam:.2f}')
ax.set_xlim(-6, 6); ax.set_ylim(-6, 6)
ax.axhline(0, color='k', lw=0.8); ax.axvline(0, color='k', lw=0.8)
ax.set_title('Eigenvectors Scaled by Eigenvalues')
ax.legend()
plt.tight_layout(); plt.show()

print("4.3 passed ✓")
````

**Your Geometric Explanation (4.3):**

>  Geometrically, when a matrix multiplies an eigenvector, the result points in the exact same (or opposite) direction as the original vector — it doesn't get rotated, only **scaled**. The eigenvalue tells us the scale factor: if |λ| > 1 the vector is **stretched**, if 0 < |λ| < 1 it is **squished**, and a negative λ flips its direction. Every other vector in the plane gets rotated *and* scaled by the matrix, but eigenvectors mark the special "axes" along which the transformation acts as pure scaling.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt = np.linalg.svd(X, full_matrices=False)  # Compute SVD

X_reconstructed = U @ np.diag(S) @ Vt  # Reconstruct X from SVD components

# Rank-1 approximation using only the largest singular value/vector
X_approx = S[0] * np.outer(U[:, 0], Vt[0, :])  # Largest singular value times outer product of corresponding singular vectors

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**

> In SVD, `X = U S Vt`: **U**'s columns are the left singular vectors (how each data point loads onto each component), **S** is a diagonal of singular values describing the magnitude/importance of each component, and **Vt**'s rows (V's columns) are the right singular vectors. When `X` is mean-centered, **V (the rows of Vt) contains the principal component directions** — the directions of maximum variance in the data — and the singular values in `S` are proportional to the square roots of the eigenvalues (variances) along those directions. So PCA can be computed directly via SVD of the centered data matrix, without ever forming the covariance matrix.
